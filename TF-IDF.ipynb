# TF-IDF

```python
import pandas as pd
```

```python
spam =pd.read_csv("spam.csv", sep=",", encoding='latin1')
spam
```

```python
spam.drop(columns=["Unnamed: 2","Unnamed: 3","Unnamed: 4"], inplace=True)
```

```python
spam.columns =["label","message"]
spam
```

### Data Cleaning and PreProcessing

```python
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import  WordNetLemmatizer
```

```python
corpus =[]
for i in range(0, len(spam)):
    sentence =spam.iloc[i, 1]
    sentence =re.sub("[^a-zA-Z]", " ", sentence)
    sentence =sentence.lower().split()
    sentence =[WordNetLemmatizer().lemmatize(word) for word in sentence  if word not in stopwords.words("english")]
    corpus.append(" ".join(sentence))
```

```python
corpus
```

### Bag of Words

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# select top 2500 words that have max count
# set binary =True for binary BOW
tf_idf =TfidfVectorizer(max_features=100)
```

```python
bow =tf_idf.fit_transform(corpus).toarray()
```
